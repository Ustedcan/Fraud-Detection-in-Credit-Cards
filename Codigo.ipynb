Para determinar si la variable `'TimeHoras'` influye en la materialización de una transacción como fraudulenta, vamos a realizar una prueba de normalidad sobre los datos para después calcular un coeficiente de correlación y verificar si existe una relación entre las variables Class, que contiene la clasificación de Fraude o No Fraude y la variable TimeHoras

Primero vamos a establecer las hipotesis:
- H0 : Los datos siguen una distribución normal
- H1 : Los datos NO siguen una distribución normal

El análisis lo realizaremos sobre los dos subsets creados: `trans_normales` y `trans_fraud`

```python
# Vamos a trabajar con los dos subsets creados anteriormente 
# Empezamos con el subset trans_fraud
# Dado que es un subset pequeño (relativamente) podemos utilizar directamente la función shapiro de Python para realizar la prueba de normalidad 

# Importar la libreria
from scipy.stats import shapiro

# Calculamos los estadisticos
statistic,p_value = shapiro(trans_fraud['TimeHoras'])
rounded_statistic = round(statistic,2)
print('El estadistico W de la prueba de Normalidad Shapiro es:', rounded_statistic)
print('El estadistico p_value de la prueba de Normalidad de Shapiro es:'"{:.20f}".format(p_value))
```

Al obtener un valor de p tan cercano a cero tenemos evidencia significativa para rechazar la hipotesis nula planteada, por lo que aceptamos la hipotesis alternativa: 
* Los datos NO siguen una distribución normal

Por lo tanto los datos para el subset de transacciones fraudulentas no sigue una distribucion normal

Para el subset transacciones `trans_normales` debido a que la prueba de Normalidad Shapiro-Wilk no es muy buena para muestras mayores a 5000 observaciones, vamos a trabajar con la herramienta de Boostrap, que es una herramienta de remuestro basada en la aleatoriedad.

```python
# Definimos una función que luego utilizaremos para generar el remuestro para cualquier variable
# Vamos a definir un numero de muestas que sera de 100 muestras con un tamano de 4000 observaciones, cada muestra sera unicas ya que la 
# funcion no tomara una muestra repetida gracias al argumento replace=False del ciclo for.
# Relicé algunas pruebas con mas de 1000 muestras y el resultado fue el mismo, por lo que un valor menor de numero de muestras es suficiente)
def bootstrap_shapiro(trans_normales, num_samples=101, sample_size=4000):
    p_values = []

# Creamos un ciclo for que itere para cada numero de muestras y que retorne el valor de p_value para cada una de esas muestras
# de esta manera podemos verificar en varias muestas si los datos del subset siguen una distribución normal basados en el resultado 
# de cada uno de los p_values de las muestras.
    for _ in range(num_samples):
        sample = np.random.choice(trans_normales, size=sample_size, replace=False)
        _, p_value = shapiro(sample)
        p_values.append(p_value)
    return p_values
```

```python
# Seteamos las opciones de visualizacion de las filas 
pd.set_option('display.max.rows', None)

# Utilizamos la funcion de remuestreo para calular los p valores en la variable TimeHoras
p_values = bootstrap_shapiro(trans_normales['TimeHoras'])

# Convertimos los p-values a una Serie de pandas para facilitar el análisis
p_values_series = pd.Series(p_values)
# Imprimimos
p_values_series
```

El resultado de la funcion de remuestreo nos da que para todas las muestras el valor de p_value es muy cercano a cero, por lo que podemos concluir que para el subset `trans_nomales` la variable TimeHoras NO sigue una distribución normal. La evidencia es significativa para rechazar la HO. 

Una vez establecido que ninguna de los subsets que componen el dataset `creditcard.csv`de la variable `TimeHoras` sigue una distribución normal, podmos calcular el coeficiente de correlacion para determinar si existen una relación entre la variable TimeHoras y la variable Class.

Debido a que se ha comprobado que los datos no siguen una distribucion normal no es posible utilizar un coeficiente que mida la dependencia lineal entre las dos variables, el coeficiente de Pearson no seria util en este caso ya que los datos que no siguen una distribución normal, por lo tanto vamos a utilizar el coeficiente de correlación de __Kendal y Spearman__
